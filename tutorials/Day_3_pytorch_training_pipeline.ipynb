```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```

```python
#numpy_arrays to pytorch tensor
import torch
X_train_tensor=torch.from_numpy(X_train).float() # Convert to float32
X_test_tensor=torch.from_numpy(X_test).float()   # Convert to float32
y_train_tensor=torch.from_numpy(y_train)
y_test_tensor=torch.from_numpy(y_test)
```

```python
X_train_tensor.shape
```

```python
# defining model
class MySimpleNN():
  def __init__(self,X):
    self.weights=torch.rand(X.shape[1],1,requires_grad=True)
    self.bias=torch.zeros(1,requires_grad=True)

  def forward(self,X):
    z=torch.matmul(X,self.weights)+self.bias
    y_pred=torch.sigmoid(z)
    return y_pred

  def loss_function(self,y_pred,y):
    # clamp prediction to avoid log(0)
    epsilon=1e-7
    y_pred=torch.clamp(y_pred,epsilon,1-epsilon)
    # calculate loss
    loss=-(y_train_tensor*torch.log(y_pred)+(1-y_train_tensor)*torch.log(1-y_pred)).mean()
    return loss


```

```python
#important parameters
learning_rate=0.1
epochs=25
```

```python
# Training pipeline
# create model
model=MySimpleNN(X_train_tensor)
# define loop
for epoch in range(epochs):
  #forword pass
  y_pred=model.forward(X_train_tensor)
  # calculate loss
  loss=model.loss_function(y_pred,y_train_tensor)
  print(f'Epoch: {epoch+1}, Loss: {loss.item()}')
  # backward pass
  loss.backward()

  # update weights
  with torch.no_grad():
    model.weights -=learning_rate*model.weights.grad
    model.bias -=learning_rate*model.bias.grad

  # zero gradients
  model.weights.grad.zero_()
  model.bias.grad.zero_()
  # print loss in each epochs
  print(f'Epoch: {epoch+1}, Loss: {loss.item()}')
```

```python
# model evaluation
with torch.no_grad():
  y_pred=model.forward(X_test_tensor)
  y_pred=(y_pred>0.5).float()
  accuracy=(y_pred==y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
